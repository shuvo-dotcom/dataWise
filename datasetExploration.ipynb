```python
import pandas as pd
weather_df = pd.read_csv('GlobalWeatherRepository.csv')
print(weather_df.head())
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

plt.figure(figsize=(10, 6))
sns.histplot(weather_df['temperature_celsius'], bins=30, kde=True)
plt.title('Temperature Distribution')
plt.show()
```

```python
top_hottest = weather_df.groupby('location_name')['temperature_celsius'].mean().nlargest(5)

plt.figure(figsize=(10, 6))
sns.barplot(x=top_hottest.values, y=top_hottest.index)
plt.title('Top 5 Hottest Locations')
plt.xlabel('Average Temperature (Â°C)')
plt.ylabel('Location')
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.histplot(weather_df['humidity'], bins=30, kde=True)
plt.title('Humidity Distribution')
plt.xlabel('Humidity (%)')
plt.ylabel('Frequency')
plt.show()
```

```python
top_precipitation = weather_df.groupby('location_name')['precip_mm'].mean().nlargest(5)

plt.figure(figsize=(10, 6))
sns.barplot(x=top_precipitation.values, y=top_precipitation.index)
plt.title('Top 5 Locations with Highest Precipitation')
plt.xlabel('Average Precipitation (mm)')
plt.ylabel('Location')
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.scatterplot(data=weather_df, x='temperature_celsius', y='humidity')
plt.title('Temperature vs Humidity')
plt.xlabel('Temperature (Â°C)')
plt.ylabel('Humidity (%)')
plt.show()
```

```python
plt.figure(figsize=(12, 8))
plt.scatter(weather_df['longitude'], weather_df['latitude'], 
            c=weather_df['temperature_celsius'], cmap='coolwarm', s=50, alpha=0.7)
plt.colorbar(label='Temperature (Â°C)')
plt.title('Global Temperature Distribution')
plt.xlabel('Longitude')
plt.ylabel('Latitude')
plt.grid(True)
plt.show()
```
